## Computing Joint Probabilities in BN

```python
def create_network(vertices: Dict[str, pd.DataFrame], edge_list: List[List[str]] = None) -> nx.DiGraph:
    g = nx.DiGraph()
    for vertex, probas in vertices.items():
        g.add_node(vertex, prob=probas)
    if edge_list:
        for edge in edge_list:
            g.add_edge(*edge)
    return g
```

```python
def merge_prob_tables(df1: pd.DataFrame, df2:pd.DataFrame, left: str, right: str) -> pd.DataFrame:
    df_first = df1.reset_index()
    first_indexes = [col for col in df_first.columns if col.startswith('index')]
    df_second = df2.reset_index()
    second_indexes = [col for col in df_second.columns if col.startswith('index')]
    
    df_first = df_first.melt(id_vars=first_indexes, 
                      var_name=f'variable_{left}',
                      value_name=f'value_{left}')
    
    df_second = df_second.melt(id_vars=second_indexes, 
                      var_name=f'variable_{right}',
                      value_name=f'value_{right}')
    
    if (f'index_{left}' not in df_first.columns) and ('index' in df_first.columns):
        df_first = df_first.rename(columns={'index': f'index_{left}'})
    if (f'index_{right}' not in df_second.columns) and ('index' in df_second.columns):
        df_second = df_second.rename(columns={'index': f'index_{right}'})
    right_index = f'index_{right}'
    if len(second_indexes)>1:
        right_index = [ind for ind in second_indexes if ind.endswith(left)][0]
                                                    
    merged = (df_first.merge(df_second, left_on=f'variable_{left}', right_on=right_index, how='inner'))
    
    return merged.drop_duplicates().reset_index(drop=True)

def get_node_state_at(g: nx.DiGraph, node: str) -> pd.DataFrame:
    predecessors = list(g.predecessors(node))
    if len(predecessors)==1:
        return get_node_state_at(g, predecessors[0])@g.nodes[node]["prob"]
    elif len(predecessors)>1:
        return pd.DataFrame(
            g.nodes[node]["prob"].T@reduce(lambda x, y: 
                   pd.DataFrame(np.outer(x, y), index=y.columns, columns=x.columns).unstack(), 
                   [get_node_state_at(g, p) for p in predecessors]
                  )
        ).T
    else:
        return g.nodes[node]["prob"]

def get_joint_probability(g: nx.DiGraph) -> pd.DataFrame:
    nodes_leaf = [node for node in g.nodes if g.out_degree(node)==0]
    nodes_head = [node for node in g.nodes if not(list(g.predecessors(node)))]
    list_of_joint_probs = []
    for nl, nh in product(nodes_head, nodes_leaf):
        list_of_joint_probs_inner = []
        traversal_paths = list(nx.all_simple_paths(g, nl, nh))
        for traversal_path in traversal_paths:
            list_of_merged_tables = []
            for parent, succ in zip(traversal_path[:-1], traversal_path[1:]):
                merged_df = merge_prob_tables(
                    g.nodes[parent]["prob"],
                    g.nodes[succ]["prob"],
                    parent, 
                    succ
                )
                list_of_merged_tables.append(merged_df)
            list_of_joint_probs_inner.append(reduce(lambda x, y: x.merge(y, how='inner'), list_of_merged_tables))
        list_of_joint_probs.append(reduce(lambda x, y: x.merge(y, how='inner'), list_of_joint_probs_inner))
    joint_prob = reduce(lambda x, y: x.merge(y, how='inner'), list_of_joint_probs).drop_duplicates()
    
    value_cols = [col for col in joint_prob.columns if col.startswith('value')]
    joint_prob['joint_probability'] = joint_prob[value_cols].prod(axis=1)
    joint_prob = joint_prob.drop(columns=[col for col in joint_prob.columns if col.startswith('index')])
    return joint_prob.drop(columns=value_cols)
```

## Synthesis of BN from observed data

```python
asia_data  = pd.read_csv(r"data/Asia.txt", sep='\t',
                          usecols=["Smoke", "Cancer", "Tuberculosis", 
                                   "Tub_or_Cancer", "Asia", "X_Ray", "Bronchitis", "dyspnea"])
asia_data = asia_data.rename(columns={
    "Smoke": "smoking",
    "Cancer": "lung_cancer",
    "Tuberculosis": "tuberculosis",
    "Tub_or_Cancer": "tub_or_cancer",
    "Asia": "visit_asia",
    "X_Ray": "x_ray",
    "Bronchitis": "bronchitis"
})
```

```python
def entropy(arr: np.array) -> np.array:
    arr_probs = arr/np.sum(arr)
    return -1*arr@np.log(arr_probs)
    

def minimum_descriptive_length(g: nx.DiGraph, data: pd.DataFrame) -> float:
    mdls = []
    n = len(data)
    for node in g.nodes:
        predecessors = list(g.predecessors(node))
        num_predecessors = 2**len(predecessors)
        card = len(data[node].unique())
        if len(predecessors):
            l_entropy = []
            for predecessor in predecessors:
                probas = data[[predecessor, node, 'dummy']].groupby([predecessor, node]).count()
                entr = sum(map(entropy, (g['dummy'] for _, g in probas.groupby(predecessor))))
                l_entropy.append(entr)
            l_entropy = sum(l_entropy)
        else:
            probas = data[[node, 'dummy']].groupby([node]).count()
            l_entropy = entropy(probas['dummy'])
        mdl = l_entropy + (card - 1)*num_predecessors*np.log(n)*0.5
        mdls.append(mdl)
    return np.round(sum(mdls), 5)
```

```python
class MDLBayesNet():
    def __init__(self, dataset: pd.DataFrame):
        self.dataset = dataset.copy()
        self.vertices = dataset.columns.tolist()
        self.dataset['dummy'] = 1.
        self.vertex_combinations = combinations(self.vertices, 2)
        self.init_graph = nx.DiGraph()
        self.init_graph.add_nodes_from(self.vertices)
        self.mi_scores = self.compute_mutual_info()
        self.models = self.fit()
    
    def compute_mutual_info(self) -> pd.DataFrame:
        mi_df = dict(variable_1=[], variable_2=[], mi=[])
        for variable_1, variable_2 in self.vertex_combinations:
            mi_df['variable_1'].append(variable_1)
            mi_df['variable_2'].append(variable_2)
            mi_df['mi'].append(mutual_info_score(self.dataset[variable_1], self.dataset[variable_2]))
        return pd.DataFrame(mi_df).sort_values(by='mi', ascending=False).reset_index(drop=True)
    
    @staticmethod
    def __generate_candidates(initial_model: nx.DiGraph, new_edge: Tuple[str]) -> List[nx.DiGraph]:
        result = []
        for edges in (new_edge, new_edge[::-1]):
            candidate = deepcopy(initial_model)
            candidate.add_edge(*edges)
            try:
                nx.find_cycle(candidate)
            except nx.NetworkXNoCycle:
                result.append(candidate)
        return result
    
    def fit(self) -> List[Tuple[float, nx.DiGraph]]:
        best_mdl = minimum_descriptive_length(self.init_graph, self.dataset)
        best_models = [(best_mdl, deepcopy(self.init_graph))]
        for _, row in tqdm(self.mi_scores.iterrows(), total=len(self.mi_scores)):
            candidate_models = []
            for _, model in best_models:
                prev_mdl = best_mdl
                candidates = self.__generate_candidates(model, row[['variable_1', 'variable_2']].values)
                if candidates:
                    mdl_candidates = [
                        (minimum_descriptive_length(c, self.dataset), c)
                        for c in candidates
                    ]
                    candidate_models.extend(mdl_candidates)
            if candidate_models:
                best_mdl = min(mdl_candidates, key=lambda x: x[0])[0]
                if (prev_mdl > best_mdl):
                    best_models = list(
                            filter(
                                lambda y: abs(y[0] - best_mdl)<1e-3,
                                candidate_models
                            )
                        )
                else:
                    best_mdl = prev_mdl
        return best_models
     
```

```python
asia_net = MDLBayesNet(asia_data)
```

```python
for mdl, g in asia_net.models:
    print(f"MDL: {mdl}, EDGE LIST: {list(g.edges)}")
```

```python
ground_truth_mdl = minimum_descriptive_length(g, asia_net.dataset)
print(f"MDL: {ground_truth_mdl}, EDGE LIST: {list(g.edges)}")
```
